```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/predicting-learning-outcomes/ds5165_student_step_All_Data_7337_2022_0107_094330.txt', sep='\t')
df
```

```python
df = df.dropna(subset=['KC (Default)', 'Opportunity (Default)'])
df
```

```python
df = df[df['Opportunity (Default)'].apply(lambda x: x.isnumeric())]
df
```

```python
df = df.rename(columns={'Anon Student Id': 'individual', 'KC (Default)': 'KC', 'Opportunity (Default)': 'opportunity', 'First Attempt': 'response'})
df
```

```python
df = df[['individual', 'KC', 'opportunity', 'response']]
df
```

```python
df['response'].unique()
```

```python
df['response'] = df['response'].map({'correct': 1, 'incorrect': 0, 'hint': 0})
df['opportunity0'] = df['opportunity'].apply(lambda x: int(x) - 1)
df
```

```python
df = df.drop(columns=['opportunity'])
df
```

```python
df.info()
```

```python
df.to_csv('ds5165_processed.txt', index=False, sep='\t')
```
